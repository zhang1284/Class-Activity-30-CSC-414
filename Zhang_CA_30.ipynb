```python
# The Current updated state object
class TrafficLightFSM:
  current_state = None
  RED = "RED"
  GREEN = "GREEN"
  YELLOW = "YELLOW"

# DECLARE the function volume_control_finite_state_machine()
def traffic_light_finite_state_machine(transition_state_tuple_list, advance_signal):
  if not advance_signal:
    return

  if TrafficLightFSM.current_state is None:
    TrafficLightFSM.current_state = transition_state_tuple_list[0][0]
    return

  transitions = dict(transition_state_tuple_list)

  if TrafficLightFSM.current_state in transitions:
      TrafficLightFSM.current_state = transitions[TrafficLightFSM.current_state]


### TEST CASES ###

transition_state_list = [
    (TrafficLightFSM.RED, TrafficLightFSM.GREEN),
    (TrafficLightFSM.GREEN, TrafficLightFSM.YELLOW),
    (TrafficLightFSM.YELLOW, TrafficLightFSM.RED),
]

advance_signal_list = [False, True, True, False, True, True]

for signal in advance_signal_list:
  traffic_light_finite_state_machine(
      transition_state_tuple_list=transition_state_list,
      advance_signal=signal
  )
  print("Traffic Light FSM Current State: {0}\n".format(TrafficLightFSM.current_state))
```
